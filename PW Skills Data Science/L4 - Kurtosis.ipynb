```python
# In this, we are doing practical of kurtosis from video lecture 4

import pandas as pd
import matplotlib.pyplot as plt

iris_df = pd.read_csv('https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv')

print(iris_df)
```

```python
# normalizing data
normalized_data = (iris_df['sepal_length'] - iris_df['sepal_length'].mean())/iris_df['sepal_length'].std()

print(normalized_data)
```

```python
# plot histogram
plt.hist(normalized_data, bins = 20, edgecolor='black')
plt.xlabel('Normalized sepal length')
plt.ylabel('Frequency')
plt.title('Histogram of Normalized sepal length')
plt.show()

# plot histogram with kurtosis greater than 3
high_kurtosis_data = (iris_df['sepal_length'] - iris_df['sepal_length'].mean())*5 # example multiply by 5 to increase kurtosis
plt.hist(high_kurtosis_data, bins=20, edgecolor='black')
plt.xlabel('High kurtosis Sepal length')
plt.ylabel('Frequency')
plt.title('Histogram with High kurtosis')
plt.show()

# plot histogram with kurtosis whose value is lesser than 3
high_kurtosis_data = (iris_df['sepal_length'] - iris_df['sepal_length'].mean())/5 # exmaple: divide by 5 to decrease kurtosis
plt.hist(high_kurtosis_data, bins=20, edgecolor='black')
plt.xlabel('High kurtosis sepal length')
plt.ylabel('Frequency')
plt.title('Histogram with high kurtosis')
plt.show()
```
