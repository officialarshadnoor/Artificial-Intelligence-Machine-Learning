```python
# startong with seaborn

# it is used for data visualization

#lets start

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

var = [1,2,3,4,5,6,7]
var_1 = [2,3,4,1,5,6,7]

x_1 = pd.DataFrame({"var":var,"var_1":var_1})
sns.lineplot(x="var", y="var_1", data=x_1)
plt.show()
```

```python
plt.plot(var,var_1)
plt.show
```

```python
# load iris.csv from local
iris = pd.read_csv("iris.csv")
```

```python
#how many data points and features?
print(iris.shape)
```

```python
# what are the column names in out dataset?
print(iris.columns)
```

```python
#how many data points for each class are present?
# or how many flowers for each species are present? 
iris["species"].value_counts()

# balanced vs imbalanced dataset
# iris is a balanced dataset as the number of data points for every class is 50.
```

```python
# 2D Scatter Plot
# always understand the axis labels and scale
iris.plot(kind='scatter', x='sepal_length',y='sepal_width')
plt.show()
```

```python
# 2d scatter plot with color coding for each flower type/class
# here 'sns' corresponds to seaborn

sns.set_style("whitegrid")
sns.FaceGrid(iris,hue="species",size=4)\
.map(plt.scatter,"sepal_length","sepal_width")\
.add_legend()
plt.show()
```
