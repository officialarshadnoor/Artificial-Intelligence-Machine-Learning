```python
# 2D scatter plot with color coding for each flower type/class
# here 'sns' corresponds to seaborn
from matplotlib import pyplot as plt
import seaborn as sns

# Load the iris dataset
iris = sns.load_dataset('iris')

sns.set_style("whitegrid")
sns.FacetGrid(iris, hue="species", height=4) \
   .map(plt.scatter, "sepal_length", "sepal_width") \
   .add_legend()
plt.show()
```

```python
# 3d scatter plot
# pair plot

# pairwise scatter plot pairplot
# disadvantages
 ## Can be used when number of features are high.
 ## Cannot visualize higher dimensional patterns in 3-D and 4-D
 ## Only possible to view 2D patterns

plt.close()
sns.set_style("whitegrid")
sns.pairplot(iris,hue="species", size=3)
plt.show()

## Note --  Here diagonal elements are PDFs which are explained below !!
```

```python
# petals length and petal width are most useful feature to identify various flower types

### Histograml, PDF, CDF

# What about 1-D scatter plot using just one feature?
# 1D scatter plot of petal length
import numpy as np
iris_setosa = iris.loc[iris["species"] == "setosa"];
iris_virginica = iris.loc[iris["species"] == "virginica"];
iris_versicolor = iris.loc[iris["species"] == "versicolor"];
#print(iris_setosa["petal_length"])
plt.plot(iris_setosa["petal_length"], np.zeros_like(iris_setosa['petal_length']), 'o')
plt.plot(iris_versicolor["petal_length"], np.zeros_like(iris_versicolor['petal_length']), 'o')
plt.plot(iris_virginica["petal_length"], np.zeros_like(iris_virginica['petal_length']), 'o')

plt.show()

# Disadvantage of 1 D scatter plot : Very hard to make sense as points
# are overlapping alot.
# Are better ways of visualizing 1 D scatter plots?
```

```python
# need for cumulative distribution function
# we can visually see what percentage of versicolor flowers have a petal length of less than 5
# how to cosntruct a CDF?
# how to read a CDF?

# Plot CDF of of petal_length

counts, bin_edges = np.histogram(iris_setosa['petal_length'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges);
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf);
plt.plot(bin_edges[1:], cdf)


counts, bin_edges = np.histogram(iris_setosa['petal_length'], bins=20, 
                                 density = True)
pdf = counts/(sum(counts))
plt.plot(bin_edges[1:],pdf);

plt.show();
```
